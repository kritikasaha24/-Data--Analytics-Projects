```python
import requests
from bs4 import BeautifulSoup
from transformers import pipeline
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter

def fetch_headlines(url):
    response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})
    return [el.get_text(strip=True) for el in BeautifulSoup(response.content, 'html.parser').find_all("span", {"itemprop": "headline"})] if response.status_code == 200 else []

def analyze_sentiments(headlines):
    return pipeline("sentiment-analysis")(headlines)

def visualize_results(headlines, sentiments):
    WordCloud(width=800, height=400, background_color='white').generate(" ".join(headlines)).to_image().show()
    plt.bar(*zip(*Counter(s['label'] for s in sentiments).items()), color=['green' if l == 'POSITIVE' else 'red' for l in sentiments])
    plt.xlabel("Sentiment"); plt.ylabel("Count"); plt.title("Sentiment Distribution"); plt.show()

def main():
    if (headlines := fetch_headlines("https://inshorts.com/en/read")):
        sentiments = analyze_sentiments(headlines)
        print(*[f"{h} -> {s}" for h, s in zip(headlines, sentiments)], sep="\n")
        visualize_results(headlines, sentiments)
    else:
        print("No headlines found.")

if __name__ == "__main__":
    main()
```

```python
import requests
from bs4 import BeautifulSoup
from transformers import pipeline
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter

def fetch_headlines(url):
    response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})
    return [el.get_text(strip=True) for el in BeautifulSoup(response.content, 'html.parser').find_all("span", {"itemprop": "headline"})] if response.status_code == 200 else []

def analyze_sentiments(headlines):
    return pipeline("sentiment-analysis")(headlines)

def visualize_results(headlines, sentiments):
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    wordcloud = WordCloud(width=400, height=400, background_color='white').generate(" ".join(headlines))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.title("Word Cloud")

    plt.subplot(1, 2, 2)
    counts = Counter(s['label'] for s in sentiments)
    plt.bar(counts.keys(), counts.values(), color=['green' if l == 'POSITIVE' else 'red' for l in counts.keys()])
    plt.xlabel("Sentiment"); plt.ylabel("Count"); plt.title("Sentiment Distribution")

    plt.tight_layout()
    plt.show()

def main():
    if (headlines := fetch_headlines("https://inshorts.com/en/read")):
        sentiments = analyze_sentiments(headlines)
        print(*[f"{h} -> {s}" for h, s in zip(headlines, sentiments)], sep="\n")
        visualize_results(headlines, sentiments)
    else:
        print("No headlines found.")

if __name__ == "__main__":
    main()

```
